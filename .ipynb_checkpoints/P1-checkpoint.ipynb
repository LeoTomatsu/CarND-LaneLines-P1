## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

# No Change
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# No Change
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

# No Change
def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

# No Change
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# New function
def best_fitting_line(lines, y1, y2):
    """
    Averages the position of each of the lines and returns points 
    extrapolate to the top and bottom of the lane.
    """
    #initialize list of x-axis and y-axis points
    x_points_list = []
    y_points_list = []
    
    
    #append points to list
    for line in lines:
        line_x1, line_y1, line_x2, line_y2 = line[0]
        
        x_points_list.append(line_x1)
        x_points_list.append(line_x2)
        y_points_list.append(line_y1)
        y_points_list.append(line_y2)

    #calculate the best-fitting slope and 
    #best-fitting y-intercept
    slope, intercept = np.polyfit(x_points_list,
                            y_points_list, 1)
    
    #determine x-axis points
    x1 = int((y1 - intercept) / slope)
    x2 = int((y2 - intercept) / slope)
    
    return x1, y1, x2, y2
            

# Changes
def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    #initialize list left and right lines 
    left_line_list = []
    right_line_list = []
    
    #store dimensions of image
    ysize = img.shape[0]
    xsize = img.shape[1]
    
    #iterate over lines to append to list of left and right lines
    for line in lines:
        for x1,y1,x2,y2 in line:
            #initialize slope value
            slope = 0
            
            #check for zero division error
            if (x2 - x1) is not None:
                slope = (y2 - y1) / (x2 - x1)
            
            #if slope is negaative and points are less than 
            #half of image size store in list of left lines
            if slope < 0 and x1 < (xsize / 2) \
                and x2 < (xsize / 2):
                    left_line_list.append(line)
            # if slope is positive and points are greater than
            # half of image size store in lsit of right lines
            elif slope > 0 and x1 > (xsize / 2) \
                and x2 > (xsize / 2):
                    right_line_list.append(line)
    
    #find best-fitting line for left and right
    left_x1, left_y1, left_x2, left_y2 = \
        best_fitting_line(left_line_list, \
        int(ysize), int(ysize / 2 + 50))
    right_x1, right_y1, right_x2, right_y2 = \
        best_fitting_line(right_line_list, \
        int(ysize), int(ysize / 2 + 50))

    #draw best-fitting line
    cv2.line(img, (left_x1, left_y1), \
        (left_x2, left_y2), color, thickness)
    cv2.line(img, (right_x1, right_y1), \
        (right_x2, right_y2), color, thickness)

# No Changes
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    
    return line_img

# Python 3 has support for cool math symbols.

# No Changes
def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory. 

# New Function    
def pipeline(image, filename, output_directory):
    '''
    Applies and returns hough transform to image.
    
    Steps of pipe listed below.
    1. Read in and grayscale the image.
    2. Apply Gaussian smoothing to image.
    3. Apply Canny edge detection to image.
    4. Apply hough transform to masked four sided polygon image.
    5. Draw the liens on edge image.
    
    '''
    image_directory = os.path.splitext(filename)[0]
    output_image_directory = os.path.join(output_directory, image_directory)
    
    # Read in and grayscale the image
    gray = grayscale(image)
    
    # plot grayscale image
    plot_image(gray, "Grayscale Image", "gray")
    
    # store grayscale image
    set_image_copy("Grayscale_Image_" + filename,
                       output_image_directory, gray, color="gray")
    
    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)
    
    plot_image(blur_gray, "Gaussian Blur Image", "gray")
    
    set_image_copy("Gaussian_Blur_Image_" + filename,
                       output_image_directory, blur_gray, color="gray")    
    
    # Define our parametesr for Canny and apply
    low_threshold = 60
    high_threshold = 180
    edges = canny(blur_gray, low_threshold, high_threshold)
    
    plot_image(edges, "Canny Edge Image", "gray")
    
    set_image_copy("Canny_Edge_Image_" + filename,
                        output_image_directory, edges, color="gray")
    
    
    # Define a four sided polygon to mask
    ysize = image.shape[0]
    xsize = image.shape[1]
    
    vertices = np.array([[
            (0, xsize),
            (xsize / 2 - 50, 3* ysize / 5),
            (xsize / 2 + 50, 3* ysize / 5),
            (xsize, ysize),
        ]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    plot_image(masked_edges, "Masked Edge Image")
    
    set_image_copy("Masked_Edge_Image_" + filename,
                        output_image_directory, masked_edges)
    
    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distnace resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 1 # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 5 # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0
    
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = hough_lines(masked_edges, rho, theta, threshold,
                                min_line_length, max_line_gap)

    plot_image(lines, "Hough Lines Image")
    
    set_image_copy("Hough_Lines_Image_" + filename,
                        output_image_directory, lines)    
    
    # Draw the lines on the edge image  
    output_image = weighted_img(lines, image)
    
    plot_image(output_image, "Hough Transform Image")
    
    #set image copy
    set_image_copy("Hough_Transform_Image_" + filename,
                       output_image_directory, output_image)
    
    return output_image

# New Function
def plot_image(image, title, color=None):
    '''
    plot given image with title and 
    '''
    plt.title(title)
    plt.imshow(image, cmap=color)
    plt.show()

# New Function
def get_image_copy(filename, filepath):
    '''
    returns image copy of filename
    '''
    # create filepath to input image
    input_image_filepath = os.path.join(filepath, filename)
    # read in the image
    image = mpimg.imread(input_image_filepath)
    # make a copy of the image
    image_copy = np.copy(image)
    
    print("DEBUG: get " + input_image_filepath)
    
    return image_copy

# New Function
def set_image_copy(filename, filepath, image, color=None):
    '''
    stores image of filename
    '''
    # create filepath to output image
    output_image_filepath = os.path.join(filepath, filename)
    # save image to output filepath
    if color == "gray":
        plt.imsave(output_image_filepath, image, cmap="gray")
    elif color is None:
        plt.imsave(output_image_filepath, image)
    print("DEBUG: set " + output_image_filepath)

# New Function    
def test_images(input_directory, output_directory):
    '''
    Apply hough transform to images in input_directory,
    plot created images to console,
    and store said images to output_directory
    '''
    print("INFO: start.")
    #applies hough transform to images in input_directory
    #and stores in output directory
    input_directory = "test_images/"
    output_directory = "test_images_output/"
    
    imagelist = os.listdir(input_directory)
    
    for i, filename in enumerate(imagelist):
        #get image copy
        image_copy = get_image_copy(filename, input_directory)
        #apply hough transform to image
        output_image = pipeline(image_copy, filename, output_directory)

    
test_images("test_images/", "test_images_output/")
print("INFO: done.")
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)

    # Read in and grayscale the image
    gray = grayscale(image)
    
    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)
    
    # Define our parameters for Canny and apply
    low_threshold = 60
    high_threshold = 180
    edges = canny(blur_gray, low_threshold, high_threshold)
    
    # Define a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[
            (0, imshape[0]),
            (imshape[1]/2 - 50,3*imshape[0]/5),
            (imshape[1]/2 + 50,3*imshape[0]/5),
            (imshape[1],imshape[0]),
        ]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
    
    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distnace resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 1 # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 5 # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0
    
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = hough_lines(masked_edges, rho, theta, threshold,
                                min_line_length, max_line_gap)

    # Draw the lines on the edge image  
    result = weighted_img(lines, image)
    
    return result

````
